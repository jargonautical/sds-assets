```python
# importing libraries (extra functions to manage http requests and JSON file formats)
import json  
import urllib.request 
```

```python
url = "http://api.open-notify.org/astros.json" # list of astronauts currently on the ISS
response = urllib.request.urlopen(url) 
result = json.loads(response.read())
result
```

```python
url = "http://api.open-notify.org/iss-now.json" # current location of the ISS
response = urllib.request.urlopen(url) 
result = json.loads(response.read())
result
```
